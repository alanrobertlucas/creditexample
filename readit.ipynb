# Read Files

```python
# Read Files
traindir='./csv_files/train/'
# thedir=os.listdir('C:/Users/AlanLucas/Documents/bootcamp/homecredit/csv_files/train/')
thedir=os.listdir(traindir)
# os.getcwd()

i=0
l=len(thedir)
for f in thedir:
    i+=1
    if i<l+1:
        thefile=traindir+f
        print(cr,"File: ",thefile,cr)
        trbase=pd.read_csv(thefile)
        #print(trbase.head(10))
        print("Dataset Shape is {}".format(trbase.shape),cr)
        # print(trbase.describe(),cr)
        # Count number of unique values and missing values
        input_uniqvals = pd.DataFrame(trbase.nunique(axis=0),columns=["Unique Values"])
        input_dtypes = pd.DataFrame(trbase.dtypes, columns=["Data Type"])
        missings=pd.DataFrame(trbase.isna().sum())
        uniq=pd.merge(left=input_uniqvals,right=input_dtypes,left_index=True,right_index=True)
        summary=pd.merge(left=uniq,right=missings,left_index=True,right_index=True)
        summary.columns.values[[2]]=['Missing']
        print(summary,cr)

```
